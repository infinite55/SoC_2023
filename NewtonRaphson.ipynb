## Newton Raphson Method

Given a function f(x) and an initial guess for root, find the root of function in interval. <br>
formula: $x_{n+1}=x_{n}-\frac{f(x_{n})}{f'(x_{n})}$ <br>


Algorithm:<br>
Input: initial x, Func(x), derivFunc(x)<br>
Output: root of Func()<br>

1. Find f'(x) and define NR equation
2. Guess an initial value of x for the first iteration
3. Substitute x in the NR equation and calculate x*
4. If $|x^{*}-x|<$ tolerance, stop iterations and output the root: x*
5. If the number of iterations reaches an assumed maximum, stop.
6. Else let x=x* and repeat the steps from 3 until a condition in 4 or 5 is satisfied.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*x**3 - 9.5*x + 7.5

def df(x):
    return 6*x**2 - 9.5

def newton_raphson(x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        dfx = df(x)
        h = fx/dfx
        xnew = x-h 
        if abs(h) < tol:
            break #print("Converged to root:", x)
            #return x
        x = xnew 
    #print("Failed to converge")
    print("The root is %f at %d iterations:" %(xnew, i))
    return xnew

# Set up plot
fig, ax = plt.subplots()
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Newton-Raphson Method')

# Plot function
x = np.linspace(-3, 3, 100)
y = f(x)
ax.grid()
ax.plot(x, y, label='f(x)')

# Plot tangent line
x0 = 0 # initial guess
y0 = f(x0)
slope = df(x0)
intercept = y0 - slope*x0
tangent_line = slope*(x - x0) + intercept
#ax.plot(x, tangent_line, linestyle='--', label='Tangent Line')

# Plot intersection point
x1 = newton_raphson(x0)
y1 = f(x1)
ax.scatter(x1, y1, color='r', label='Root')

x0 = 5 # initial guess
y0 = f(x0)

x2 = newton_raphson(x0)
y2 = f(x2)
ax.scatter(x2, y2, color='b', label='Root')

x0 = -4 # initial guess
y0 = f(x0)

x3 = newton_raphson(x0)
y3 = f(x3)
ax.scatter(x3, y3, color='y', label='Root')

# Show plot
ax.legend()
plt.show()
```
