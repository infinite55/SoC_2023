```python
import turtle
import math
import random
```

```python
# Set up the canvas and turtle
canvas = turtle.Screen()
canvas.title("Random Walk")
canvas.bgcolor("white")
canvas_width = 800
canvas_height = 800

turtle.speed(0)
turtle.penup()
turtle.goto(0, 0)
turtle.pendown()

dist = canvas_height / 2
turtle.goto(-dist, -dist)
turtle.goto(-dist, dist)
turtle.goto(dist, dist)
turtle.goto(dist, -dist)
turtle.goto(-dist, -dist)

posX = 0
posY = 0

def random_walk():
    global posX, posY

    dir = math.floor(random.random() * 2)
    sense = math.floor(random.random() * 2)

    if dir == 0:
        posX += (1 if sense else -1) * 4
    else:
        posY += (1 if sense else -1) * 4

    turtle.goto(posX, posY)

    if abs(posX) > dist or abs(posY) > dist:
        posX = 0
        posY = 0
        turtle.clear()
        turtle.goto(-dist, -dist)
        turtle.goto(-dist, dist)
        turtle.goto(dist, dist)
        turtle.goto(dist, -dist)
        turtle.goto(-dist, -dist)
        
    # Call random_walk() again after 50 milliseconds
    turtle.ontimer(random_walk, 50)

# Start the random walk
random_walk()

# Close the canvas on click
canvas.exitonclick()
```
