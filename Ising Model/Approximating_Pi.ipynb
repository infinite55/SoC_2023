```python
import random

total_count = 10**7 
inside_count = 0

for i in range(total_count):
    x = random.random()
    y = random.random()
    if x**2 + y**2 < 1:
        inside_count += 1

pi = inside_count * 4.0 / total_count
print(pi)
```

Above code approximates the value of $\pi$ using a Monte Carlo method. It generates random points within the unit square and checks if they fall within the quarter circle's area. The ratio of points inside the circle to the total points gives an approximation of $\pi/4$, which is then multiplied by 4 to obtain an approximation of $\pi$.

```python
import turtle
import random
```

```python
canvas = turtle.Screen()
canvas.title("Approximating Pi using Monte Carlo")
canvas.bgcolor("white")
radius = 250
turtle.speed(0)
turtle.penup()
turtle.goto(0, -radius)
turtle.pendown()
turtle.hideturtle()

totalPoints = 0
insidePoints = 0
running = True  # Flag to control the loop

def randomPoint():
    global totalPoints, insidePoints

    x = 2. * random.random() - 1.
    y = 2. * random.random() - 1.

    p = x * x + y * y
    if p < 1:
        turtle.color("red")
        insidePoints += 1
    else:
        turtle.color("black")
    
    totalPoints += 1
    turtle.penup()
    turtle.goto(x * radius, y * radius)
    turtle.pendown()
    turtle.dot()

    if totalPoints % 100 == 0:
        piApprox = insidePoints / totalPoints * 4.
        print(piApprox)
        if round(piApprox, 2) == 3.14:
            totalPoints = 0
            insidePoints = 0
            turtle.clear()
            
    # Call the randomPoint function again after 10 milliseconds
    canvas.ontimer(randomPoint, 10)

def stop_simulation(x, y):
    global running
    running = False

# Bind the stop_simulation function to mouse click
canvas.onclick(stop_simulation)

# Start the simulation
randomPoint()

# Close the canvas on click
canvas.exitonclick()
```

The code generates points at 'random' positions within the square and tallies the points that fall within the circular region. The approximation of $\pi$ is calculated by comparing the count of points inside the circle to the total number of generated points. Once the approximation reaches 3.14, the process restarts.
